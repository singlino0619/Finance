```python
import pandas
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Box_Muller_method(num):
    out = []
    for i in range(num):
        X = np.random.rand()
        Y = np.random.rand()
        rand_normal_dist = np.sqrt(- 2 * np.log(X)) * np.cos(2*np.pi*Y)
        out.append(rand_normal_dist)
    return out
```

```python
figure = plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.hist(Box_Muller_method(100), bins=50)
plt.grid(True)
plt.subplot(2,2,2)
plt.hist(Box_Muller_method(1000), bins=50)
plt.grid(True)
plt.subplot(2,2,3)
plt.hist(Box_Muller_method(10000), bins=50)
plt.grid(True)
plt.subplot(2,2,4)
plt.hist(Box_Muller_method(100000), bins=50)
plt.grid(True)
```
